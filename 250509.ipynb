enumerate() 함수 (내부함수) enumerate() 함수는 iterable 객체(리스트, 튜플, 문자열 등)를 입력받아 인덱스와 값을 함께 반환하는 iterator 객체를 생성 (리스트의 각 항목에 대해 순서대로 번호를 매겨주는 역할) 입력은 주로 매개변수,

기본적인 사용법 my_list = ['apple', 'banana', 'orange']

for index, value in enumerate(my_list): print(index, value)

```python
import matplotlib.pyplot as plt
```

```python
m = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
r = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize=(12,6))

# enumerate() 함수를 사용하여 반복문을 간소화
for i, month in enumerate(m):
    plt.annotate(r[i], (month, r[i]), xytext=(5, 5), textcoords='offset points')

plt.plot(m, r, color='y', marker='s', linestyle = '-.')
plt.title('average monthly rainfall')
plt.xlabel('m')
plt.ylabel('r')
plt.show()
```

```python
print(np.zeros((2, 3)))      # 2x3 크기의 0 배열
print(np.ones(5))            # 길이 5의 1 배열
print(np.arange(0, 10, 2))   # 0부터 10까지 2씩 증가
print(np.linspace(0, 1, 5))  # 0~1 사이를 5개로 나눈 배열
print(np.random.rand(3))     # 0~1 사이의 난수 3개
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Shape:", arr.shape)    # (2, 3)
print("Data Type:", arr.dtype)  # int64
print("Size:", arr.size)      # 6
print("Dimension:", arr.ndim) # 2
```
